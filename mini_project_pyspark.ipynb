**Basic pyspark project**

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .appName("ExcelReader") \
    .getOrCreate()
```

```python
df = spark.read.csv("/content/orders(1).csv", header=True, inferSchema=True)
df.show()
```

```python
df.show(5)
```

```python
df.printSchema()
```

```python
from pyspark.sql.functions import *
```

```python
df_distinct = df.distinct()
```

```python
df_distinct.show()
```

**dropDuplicates()**

```python
df_unique_customers = df.dropDuplicates(["Customer_ID"])
```

```python
df_unique_customers.show()
```

**withColumnRenamed()**

```python
df1 = df.withColumnRenamed("Order_Date", "order_date")
```

```python
df1.show()
```

**withColumn**

```python
df2 = df1.withColumn("Source", lit("Online"))
```

```python
df2.show()
```

```python
df3 = df2.withColumn("order_year", year(to_date(col("order_date"), "dd-MM-yyyy")))
```

```python
df3.show()
```

```python
df4 = df3.withColumn("adjusted_quantity", col("Quantity") + 10)
```

```python
df4.show()
```

**filter()**

```python
df_filtered1 = df.filter(col("Quantity") > 3)
```

```python
df_filtered1.show()
```

```python
df_filtered4 = df.filter((col("Quantity") > 3) & (col("Occasion") == "Birthday"))
```

```python
df_filtered4.show()
```

**sort()**

```python
df_sorted = df.sort("Quantity")
```

```python
df_sorted.show()
```

**orderby**


```python
df_ordered = df.orderBy(col("Quantity").desc())
```

```python
df_ordered.show()
```

**groupby**

```python
from pyspark.sql.functions import count, sum
```

```python
orders_by_location = df.groupBy("Location").agg(count("Order_ID").alias("Total_Orders"))
```

```python
orders_by_location.show()
```

**Union and Union all**


```python
df_a = df.select("Customer_ID", "Location").limit(5)
df_b = df.select("Customer_ID", "Location").limit(5)
df_union_all = df_a.union(df_b)
df_union = df_a.union(df_b).distinct()
```

```python
df_union_all.show()
```

```python
df_union.show()
```

**pivot**


```python
pivot_df = (
    df.groupBy("Location")
      .pivot("Occasion")
      .agg(sum("Quantity"))
)

```

```python
pivot_df.show()
```

```python
from pyspark.sql.functions import expr
```

```python
unpivot_df = pivot_df.select(
    "Location",
    expr("stack(3, 'Anniversary', Anniversary, 'Birthday', Birthday, 'Diwali', Diwali) as (Occasion, Total_Quantity)")
)

```

```python


unpivot_df.show()
```

```python
from pyspark.sql.functions import date_format
```

```python
df = df.withColumn("order_date_excel", date_format("order_date", "yyyy-MM-dd"))



```

```python
df.show()
```

**Window**


```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import row_number
from pyspark.sql.window import Window

spark = SparkSession.builder.appName("SimpleWindow").getOrCreate()
df = spark.read.option("header", "true").csv("/content/orders(1).csv", inferSchema=True)
windowSpec = Window.partitionBy("Customer_ID").orderBy("Quantity")
df = df.withColumn("row_num", row_number().over(windowSpec))

df.show()
```
